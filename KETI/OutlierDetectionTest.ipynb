# outlierDetection Class Test 

```python
import os
import pandas as pd
import outlierDetection as OD
```

## Data Preparation

```python
# 해당 코드는 CSV 입력을 기준으로 하나, CSV 를 포함한 다양한 인풋이 활용될 수 있음
train_data_path= '../data/nasa_bearing_train.csv'
test_data_path= '../data/nasa_bearing_test.csv'
# Get Train and Test Data

train = pd.read_csv(train_data_path, index_col=0, parse_dates=[0])
test = pd.read_csv(test_data_path, index_col=0, parse_dates=[0])
```

## Plot function to check test result

```python
def plotOutlierDetectionScore(train, test, method, parameter = None):
    OutD = OD.OutlierDetection()
    OutD.setMethod(method, parameter)
    model = OutD.trainModel(train)
    score = OutD.getScore(model, test)
    score.plot()
```

## Test Algorithms

```python
# KDE
method ='KDE' 
plotOutlierDetectionScore(train, test, method)
```

```python
#IForest
method ='IForest'
plotOutlierDetectionScore(train, test, method)
```

```python
#LOF
method ='LOF'
parameter = {'n_neighbors':5}
plotOutlierDetectionScore(train, test, method, parameter)
```
